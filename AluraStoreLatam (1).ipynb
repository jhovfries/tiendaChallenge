### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
##precios totales
total_precio_tienda1 = tienda['Precio'].sum()
total_precio_tienda2 = tienda2['Precio'].sum()
total_precio_tienda3 = tienda3['Precio'].sum()
total_precio_tienda4 = tienda4['Precio'].sum()

print(f"Total ingresos tienda 1: {total_precio_tienda1:,}")
print(f"Total ingresos tienda 2: {total_precio_tienda2:,}")
print(f"Total ingresos tienda 3: {total_precio_tienda3:,}")
print(f"Total ingresos tienda 4: {total_precio_tienda4:,}")
```

# 2. Ventas por categoría

```python
ventas_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

```python
ventas_categoria = pd.DataFrame({
    'Tienda 1': ventas_tienda1,
    'Tienda 2': ventas_tienda2,
    'Tienda 3': ventas_tienda3,
    'Tienda 4': ventas_tienda4
}).fillna(0).astype(int)
```

# 3. Calificación promedio de la tienda


```python
calificacion_prom = tienda['Calificación'].mean()
calificacion_prom2 = tienda2['Calificación'].mean()
calificacion_prom3 = tienda3['Calificación'].mean()
calificacion_prom4 = tienda4['Calificación'].mean()
```

```python
print(f"Calificación promedio tienda 1: {calificacion_prom:.4f}")
print(f"Calificación promedio tienda 2: {calificacion_prom2:.4f}")
print(f"Calificación promedio tienda 3: {calificacion_prom3:.4f}")
print(f"Calificación promedio tienda 4: {calificacion_prom4:.4f}")
```

# 4. Productos más y menos vendidos

```python
def productos_mas_menos_vendidos(df, n):
    ventas_por_producto = df.groupby('Producto')['Precio'].count().sort_values(ascending=False)
    mas_vendidos = ventas_por_producto.head(5)
    menos_vendidos = ventas_por_producto.tail(5)
    print(f"\nTienda {n} más vendidos:")
    print(mas_vendidos)
    print(f"\nTienda {n} menos vendidos:")
    print(menos_vendidos)
```

```python
productos_mas_menos_vendidos(tienda, 1)
productos_mas_menos_vendidos(tienda2, 2)
productos_mas_menos_vendidos(tienda3, 3)
productos_mas_menos_vendidos(tienda4, 4)
```
